```python
import numpy 
```

```python
from sklearn.datasets.samples_generator import make_blobs
```

```python
import matplotlib.pyplot as plt
```

```python
def dist(v1, v2):
    return np.sum(np.abs(v1 - v2))
```

```python
def argmin(lst):
  return lst.index(min(lst))
```

```python
def mykmeans(X, k):
    centers = np.random.uniform(low=X.min(), high=X.max(), size=(k, X.shape[1]))
    ypr = np.ones((X.shape[0],), dtype=float)
    new_ypr = np.zeros((X.shape[0],), dtype=float)
    epoch = 0

    
   
    while((new_ypr != ypr).any()):
        ypr = np.copy(new_ypr)
        new_ypr = np.asarray(
            [argmin([dist(X[j], centers[i]) for i in range(centers.shape[0])]) 
             for j in range(X.shape[0])])
        
        for i in range(k):
            if(new_ypr[new_ypr == i].shape[0] != 0):
                centers[i] = np.mean(X[new_ypr == i], axis=0)
            else:
                centers[i] = np.random.uniform(low=X.min(), high=X.max(), 
                                               size=(X.shape[1],))
        epoch += 1
      
        
    return new_ypr, centers, epoch
```

```python
for k in range(3,7):
    for j in range(1, 100, 25):
        X, y = make_blobs(n_samples=500, centers=5, n_features=2, random_state=j) 
        ypr, centers, epoch = mykmeans(X, k)

        plt.scatter(X[ypr == 0][:,0], X[ypr == 0][:,1], c = 'blue')
        plt.scatter(X[ypr == 1][:,0], X[ypr == 1][:,1], c = 'yellow')
        plt.scatter(X[ypr == 2][:,0], X[ypr == 2][:,1], c = 'white')
        plt.scatter(X[ypr == 3][:,0], X[ypr == 3][:,1], c = 'pink')
        plt.scatter(X[ypr == 4][:,0], X[ypr == 4][:,1], c = 'green')
        plt.scatter(X[ypr == 5][:,0], X[ypr == 5][:,1], c = 'orange')
        plt.scatter(X[ypr == 6][:,0], X[ypr == 6][:,1], c = 'black')


        plt.scatter(centers[:,0], centers[:,1], c = 'red', marker="*", s = 200, 
                    label = "k=" + str(k) + ", iter=" + str(epoch))
        plt.legend(loc = 3)
        plt.show()
```

```python
X, y = make_blobs(n_samples=500, centers=5, n_features=2, random_state=100) 
ypr, centers, epoch = mykmeans(X, 5)

plt.scatter(X[ypr == 0][:,0], X[ypr == 0][:,1], c = 'blue')
plt.scatter(X[ypr == 1][:,0], X[ypr == 1][:,1], c = 'yellow')
plt.scatter(X[ypr == 2][:,0], X[ypr == 2][:,1], c = 'white')
plt.scatter(X[ypr == 3][:,0], X[ypr == 3][:,1], c = 'pink')
plt.scatter(X[ypr == 4][:,0], X[ypr == 4][:,1], c = 'green')
plt.scatter(X[ypr == 5][:,0], X[ypr == 5][:,1], c = 'orange')
plt.scatter(X[ypr == 6][:,0], X[ypr == 6][:,1], c = 'black')


plt.scatter(centers[:,0], centers[:,1], c = 'red', marker="*", s = 200, label = "k=" + str(k) + ", iter=" + str(epoch))
plt.legend(loc = 3)
plt.show()
```
